```python
#Import all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read data from the challenge_dataset
dataframe = pd.read_csv('input/itemsale.csv')
```

```python
print (dataframe.shape[1])
dataframe.columns = range(dataframe.shape[1])
print (dataframe)
```

```python
print ("Dimension of data {}".format(dataframe.shape))

print ("Basic statistical description:")
dataframe.describe()

df=dataframe.values
labels = df[:,0]
#sizes = dataframe[[1]].values
sizes = df[:,1]
print (labels)
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes,  labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
result = []
itemId = 1
for num in sizes:
    x = np.repeat(itemId, num)
    result = np.concatenate((result, x))
    itemId = itemId + 1
```
